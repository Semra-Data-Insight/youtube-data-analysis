```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:\\Users\\semra\\OneDrive\\Desktop\\youtube_data.csv")
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["description"]=df["description"].fillna("Unknown")
```

```python
df["hashtags"]=df["hashtags"].fillna("None")
```

```python
category_views=df.groupby("category")["views"].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x=category_views.index,y=category_views.values,palette="viridis")
plt.title("Kategorilere Göre Ortalama İzlenme Sayısı",fontsize=14)
plt.xlabel("Ortalama İzlenme")
plt.ylabel("Kategori")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
numeric_columns=df[["duration","bitrate","bitrate(video)","height","width","frame rate","views","likes","comments"]]
corelation_matrix=numeric_columns.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corelation_matrix,annot=True,fmt=".2f",cmap="coolwarm",square=True)
plt.title("Sayısal Değişkenler Arası Koralasyon Matrisi", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
sns.scatterplot(data=df,x="duration",y="likes",alpha=0.6)
plt.title("Süre vs Beğeni")
plt.xlabel("Süre (saniye")
plt.ylabel("Beğeni")
plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x="duration", y="comments", alpha=0.6)
plt.title("Süre vs Yorum")
plt.xlabel("Süre (saniye)")
plt.ylabel("Yorum")
```

```python
category_stats=df.groupby("category")[["likes","views"]].mean().sort_values(by="likes",ascending=False)
```

```python
category_stats
```

```python
category_melted = category_stats.reset_index().melt(id_vars="category", value_vars=["likes", "views"], var_name="Metric", value_name="Value")
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=category_melted, x="Value", y="category", hue="Metric", palette=["skyblue", "orange"])

plt.xscale("log")
plt.xlabel("Ortalama Değer (Log Ölçek)")
plt.ylabel("Kategori")
plt.title("Kategorilere Göre Ortalama Beğeni ve İzlenme (Log Ölçek)")
plt.legend(title="Metrik")
plt.tight_layout()
plt.show()
```

```python
filtered_df=df[df["views"]>=100].copy()
```

```python
filtered_df["engagement_rate"]=df["likes"]/df["views"]
```

```python
category_engagement=filtered_df.groupby("category")["engagement_rate"].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x=category_engagement.values,y=category_engagement.index,palette="coolwarm")
plt.xlabel("Ortalama Beğeni/İzlenme Oranı")
plt.ylabel("Kategorilere Göre Ortalama İzlenme Oranı")
plt.tight_layout()
plt.show()
```

```python
filtered=df[df["views"]>=100].copy()
```

```python
filtered=filtered[filtered["views"]>0]
```

```python
filtered["comment_rate_pct"]=(filtered["comments"]/filtered["views"])*100
```

```python
category_comment=(
    filtered.groupby("category")["comment_rate_pct"].mean().sort_values(ascending=False)
)
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x=category_comment.values,y=category_comment.index)
plt.xlabel("Ortalama Yorum / İzlenme Sayısı (%)")
plt.ylabel("Kategori")
plt.title("Kategorilere Göre Ortalama Yorum / İzlenme Oranı (views≥ 100")
plt.tight_layout()
plt.show()
```

```python
bins=[0,300,1200,filtered["duration"].max()]
labels=["Kısa (0-5 dk)","Orta (5-20 dk)","Uzun (20+dk)"]
filtered["duration_group"]=pd.cut(filtered["duration"],bins=bins,labels=labels,include_lowest=True)
```

```python
filtered["duration_group"]
```

```python
duration_stats=filtered.groupby("duration_group")[["views","likes","comments"]].mean()
```

```python
duration_stats
```

```python
duration_stats_plot=duration_stats[["likes","comments"]].reset_index().melt(id_vars="duration_group",var_name="Metric",value_name="Average")
```

```python
duration_stats_plot
```

```python
plt.figure(figsize=(10,8))
sns.barplot(data=duration_stats_plot,x="duration_group",y="Average",hue="Metric",palette=["skyblue","orange"])
plt.title("Video Süre Gruplarına Göre Ortalama Beğeni ve Yorum")
plt.xlabel("Video Süresi Grubu")
plt.ylabel("Ortalama Değer")
plt.legend(title="Metrik")
plt.tight_layout()
```

```python
filtered=df[df["views"]>=100].copy()
```

```python
bins=[0,300,1200,filtered["duration"].max()]
labels=["Kısa (0-5 dk)", "Orta (5-20 dk)", "Uzun (20+ dk)"]
```

```python
filtered["duration_group"]=pd.cut(filtered["duration"],bins=bins,labels=labels,include_lowest=True)
```

```python
view_stats=filtered.groupby("duration_group")["views"].mean()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=view_stats.index,y=view_stats.values,color="orange")
plt.title("Video Süre Gruplarına Göre Ortalama İzlenme")
plt.ylabel("Ortalama İzlenme Sayısı")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x=view_stats.index,y=view_stats.values,marker="o",color="orange")
plt.title("Video Süre Gruplarına Göre Ortalama İzlenme")
plt.ylabel("Ortalama İzlenme Sayısı")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=view_stats.index,y=view_stats.values,color="orange")
sns.lineplot(x=view_stats.index,y=view_stats.values,marker="o",color="blue")
plt.title("Video Süre Gruplarına Göre Ortalama İzlenme")
plt.ylabel("Ortalama İzlenme Sayısı")
plt.tight_layout()
plt.show()
```

```python
like_comment_stats = (
    filtered.groupby("duration_group")[["likes", "comments"]]
    .mean()
    .reindex(labels)  # Etiket sırasını koru
)

# Likes için combo chart
plt.figure(figsize=(8,5))
sns.barplot(x=like_comment_stats.index, y=like_comment_stats["likes"], color="skyblue")
sns.lineplot(x=like_comment_stats.index, y=like_comment_stats["likes"], marker="o", color="blue")
plt.title("Video Süre Gruplarına Göre Ortalama Beğeni")
plt.xlabel("Video Süresi Grubu")
plt.ylabel("Ortalama Beğeni")
plt.tight_layout()
plt.show()
```
